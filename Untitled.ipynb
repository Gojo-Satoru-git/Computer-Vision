```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
puppy = cv2.imread('puppy.jpg')
```

```python
puppy = cv2.cvtColor(puppy,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(puppy,cmap = 'grey')
plt.axis('off')
plt.show()
```

**Histogram shifting**

```python
def shift_hist(image,shift):
    shifted_image = image.copy().astype(np.int16)
    shifted_image = np.clip((shifted_image + shift),0,255)
    return shifted_image.astype(np.uint8)
    
```

```python
fig,ax = plt.subplots(3,2,figsize =(10,14))
img_list = [puppy,shift_hist(puppy,50),shift_hist(puppy,-100)]
img_title = ['original','shifted by 50','shifted by -50']
for i in range(3):
    ax[i,0].imshow(img_list[i],cmap = 'grey')
    sns.histplot(img_list[i].flatten(),kde = True , ax=ax[i,1])
    ax[i,0].set_title(img_title[i])
    ax[i,1].set_title(f'Histogram and kde of {img_title[i]}')
    ax[i,0].axis('off')
    
    
```

**Contrast streching**

```python
def contrast_stretch(img):
    img = ((img - np.min(img))/(np.max(img) - np.min(img))) * 255
    return img
```

```python
puppy_contrast_stretched = contrast_stretch(puppy)
img_list = [puppy,puppy_contrast_stretched]
img_title = ['Original','contrast_stretched']
fig,ax = plt.subplots(2,2,figsize =(10,8))
for i in range(2):
    ax[i,0].imshow(img_list[i],cmap = 'gray')
    sns.histplot(img_list[i].flatten(),kde = True,ax = ax[i,1])
    ax[i,0].set_title(img_title[i])
    ax[i,0].axis('off')
    
```

**Threshold limiting**

```python
def Threshold_limit(img,threshold):
    resultant_img = np.zeros_like(img)
    resultant_img[img > threshold] = 255
    return resultant_img
```

```python
img_list = [puppy,Threshold_limit(puppy,np.median(puppy)),Threshold_limit(puppy,np.median(puppy)/2),Threshold_limit(puppy,np.median(puppy)+20)]
img_title = ['Original','Threshold = median','Threshold = median/2','Threshold = median + 20']
fig,ax = plt.subplots(4,2,figsize =(10,8))
for i in range(4):
    ax[i,0].imshow(img_list[i],cmap = 'gray')
    sns.histplot(img_list[i].flatten(),kde = True,ax = ax[i,1])
    ax[i,0].set_title(img_title[i])
    ax[i,0].axis('off')
    
```

**LINEAR FILTERS**

*SMOOTHENING OR BLUR*

```python
def smoothit(img):
    resultant_img = np.zeros_like(img)
    h, w = img.shape
    img = np.pad(img,pad_width=1,constant_values = 0)
    for i in range(1,h+1):
        for j in range(1,w+1):
            win = img[i-1:i+2,j-1:j+2]
            temp = np.sum(win * (1/9))//9
            resultant_img[i-1,j-1] = np.clip(temp,0,255)        
    return resultant_img
```

```python
img_list = [puppy,smoothit(puppy)]
img_title = ['Original','Smoothened image']
fig,ax = plt.subplots(2,2,figsize =(10,8))
for i in range(2):
    ax[i,0].imshow(img_list[i],cmap = 'gray')
    sns.histplot(img_list[i].flatten(),kde = True,ax = ax[i,1])
    ax[i,0].set_title(img_title[i])
    ax[i,0].axis('off')
    
```

**Sharpening**

```python
def sharpenit(img):
    T = np.array([[0,0,0],[0,2,0],[0,0,0]])-1/9 
    resultant_img = np.zeros_like(img)
    h, w = img.shape
    img = np.pad(img,pad_width=1,constant_values = 0)
    for i in range(1,h+1):
        for j in range(1,w+1):
            win = img[i-1:i+2,j-1:j+2]
            temp = np.mean(win * T)
            resultant_img[i-1,j-1] = max(0,min(temp,255))
    return resultant_img
```

```python
img_list = [puppy,sharpenit(puppy)]
img_title = ['Original','sharpened image']
fig,ax = plt.subplots(2,2,figsize =(10,8))
for i in range(2):
    ax[i,0].imshow(img_list[i],cmap = 'gray')
    sns.histplot(img_list[i].flatten(),kde = True,ax = ax[i,1])
    ax[i,0].set_title(img_title[i])
    ax[i,0].axis('off')
```

**DFT**

```python

fourier = cv2.dft(np.float32(puppy), flags=cv2.DFT_COMPLEX_OUTPUT)
fourier_shift = np.fft.fftshift(fourier)
magnitude = 20*np.log(cv2.magnitude(fourier_shift[:,:,0],fourier_shift[:,:,1]))
magnitude = cv2.normalize(magnitude, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8UC1)
plt.imshow(magnitude,cmap = 'gray')
```
